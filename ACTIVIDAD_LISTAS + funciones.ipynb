````python
Comida = ["combo 1","combo 2", "combo 3", "combo 4","papas","jugos naturales","cerveza","agua","gaseosas"]
precios=[20000,25000,18000,24500,3000,7000,3500,4000,2500]

for menu in range(len(Comida)):
  print(Comida[menu],"-", precios[menu])

a=0
r="s"
user="s"
while user=="s":
 while r=="s":
  x=input("Qué desea?")
  if x in comida:
    cantidad=int(input(f"Cuántas unidades de {x} quieres pedir: "))
    a=a+precios[comida.index(x)]*cantiad
  else:
     x=input("Desea algo más?")
     if x=="s": continue
     else: break
  user=input("Hay mas usuarios? responda sí o no: ")
print(f"Usted ha ganado ${a} en el día")
````

#FUNCIÓN
Es un pedazo de código que se ejecuta aparte, a veces se conoce como copiado dinámico

Ventajas de implementar las funciones:
1. Vuelve el código más limpio
2. Elimina el código espaguetti
3. Establece un orden de búsqueda
4. Permite la reutilización del código

sintaxis.


```
def (nombre_de_la_función):
  Aquí va todo lo que desees ejecutar
```

Existen 4 tipos de funciones:

1. sin parametros, sin retorno
2. Sin parametros, con retorno
3. con parametros, sin retorno
4. con parametros, con retorno
