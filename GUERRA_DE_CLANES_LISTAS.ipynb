```python
votos = [0, 0, 0, 0, 0]
n = int(input("cuantos votarán: "))
for i in range (n):
  voto = int(input("vote por su candidato"))
  if  voto >= 1 and voto <= 5:
    votos[voto-1]+= 1
  else:
    print("voto invalido")

print("resultados primera vuelta: ")
for i in range(5):
  print("candidato", i+1, ":", votos[i], "votos")

primero = votos.index(max(votos))
votos_temp= votos.copy()
votos_temp[primero]= -1
segundo = votos.index(max(votos_temp))
print("Pasan a segunda vuelta:")
print("Candidato", primero+1)
print("Candidato", segundo+1)


votos2 = [0, 0]

print("Segunda vuelta")

for i in range(n):
 voto = int(input(f"Vote por {primero+1} o {segundo+1}: "))

if voto == primero+1:
 votos2[0] += 1
elif voto == segundo+1:
 votos2[1] += 1
else:
 print("Voto invalido")

# resultados finales
print("Resultados finales:")

print("Candidato", primero+1, ":", votos2[0], "votos")
print("Candidato", segundo+1, ":", votos2[1], "votos")

if votos2[0] > votos2[1]:
 print("Ganador: Candidato", primero+1)
else:
 print("Ganador: Candidato", segundo+1)
```
