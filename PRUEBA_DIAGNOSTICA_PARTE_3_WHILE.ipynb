##TABLAS DE MULTPLICAR

```python
numero = int(input("Ingresa un número: "))
i = 1

while i <= 10:
    print(numero, "x", i, "=", numero * i)
    i += 1
```

##SUMA DE NUMEROS NATURALES HASTA N

```python
n = int(input("Ingresa un número: "))
i = 1
suma = 0

while i <= n:
    suma += i
    i += 1

print("La suma es:", suma)
```

##FACTORIAL DE UN NÚMERO

```python
n = int(input("Ingresa un número: "))

if n < 0:
    print("No existen factoriales negativos")
else:
    factorial = 1
    i = 1

    while i <= n:
        factorial *= i
        i += 1

    print("El factorial es:", factorial)
```

##SERIE FIBONACCI

```python
n = int(input("Cuántos números de Fibonacci quieres: "))

a = 0
b = 1
i = 1

while i <= n:
    print(a)
    c = a + b
    a = b
    b = c
    i += 1
```

##ADIVINAR UN COLOR

```python
colores = []

i = 0
while i < 10:
    color = input("Ingresa un color: ")
    colores.append(color)
    i += 1

secreto = input("Color secreto: ")

intentos = 1
encontrado = False

while intentos <= 10 and not encontrado:
    intento = input("Adivina el color: ")

    if intento == secreto:
        print("Adivinaste")
        encontrado = True
    else:
        print("Incorrecto")



    intentos += 1
```
