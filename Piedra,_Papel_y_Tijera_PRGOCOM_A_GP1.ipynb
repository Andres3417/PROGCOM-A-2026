#**Juego simple de piedra papel y tijera en phyton.**



```python
import random

print("Elige una de las opciones: ")
print("1 = Piedra")
print("2 = Papel")
print("3 = Tijera")

x = int(input("Escribe tu número: "))
bot = random.randint(1,3)

if bot == 1:
 print("el bot eligió piedra")
elif bot == 2:
 print("el bot eligió papel")
else:
  print("el bot eligió tijera")

if x == bot:
  print("empate")
elif x == 1 and bot == 3:
  print("Ganaste")
elif x == 2 and bot == 1:
  print("Ganaste")
elif x == 3 and bot == 2:
  print("Ganaste")
else:
  print("perdiste")
```
